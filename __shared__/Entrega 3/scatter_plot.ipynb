```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../resultados.csv')
df_filt = df[df['season_end_year']>=2020]
df_filt
```

```python
lista_ano = []
for i,ano in enumerate(df_filt['season_end_year'].unique()):
    lista_ano.append(ano)


dic ={}
for i,time in enumerate(df_filt['team'].unique()):
    lista_metrica = [0]*5
    dic[time] = lista_metrica
    for j,position in enumerate(df_filt[df['team']==time]['position']):
        dic[time][j] = 21-position

# print(dic)

dic2 = {}
for k, v in dic.items():
    dic2[k] = float(sum(v)) / len(v)

dic2_sorted = dict(sorted(dic2.items(), key=lambda item: item[1], reverse=True))
print(dic2_sorted)
```

```python
# Carregar os dados de premier-league.csv
df_premier = pd.read_csv('../premier-league.csv')

# Obter os nomes dos times únicos em cada arquivo
times_premier = df_premier['club_name'].unique()
times_resultados = df['team'].unique()

# Printar os nomes dos times em ordem alfabética
print("Times em premier-league.csv (ordenados):")
print(np.sort(times_premier))

print("\nTimes em resultados.csv (ordenados):")
print(np.sort(times_resultados))
```

```python
df_premier['desempenho'] = 0
```

```python
for i, time in enumerate(df_premier['club_name']):
    if time in dic2_sorted:
        df_premier['desempenho'][i] = dic2_sorted[time]
    else:
        df_premier['desempenho'][i] = 0
```
